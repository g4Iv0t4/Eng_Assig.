```python
# "Not be concerned about how the engine gets filled with air, nor gets rid of exhaust gas, nor of any scavenging during the valve overlap period." Design and Simulation of four-Stroke Engines

# ENGINE PARAMETERS
RPM = 1000
RPM_RANGE = [1000, 1500, 1750, 2000, 2500, 3000, 3500, 4000, 4250, 4500, 5000, 5500, 6000, 6500]
bore = 0.079                                                                                 # m^2
stroke = 0.0814                                                                              # m
conn_rod = 0.137                                                                             # m
gudgeon_offset = 0                                                                           # m
crank_radius = stroke / 2                                                                    # m
CR = 11.16                                                                                   # Compression ratio
n_sigma = 4                                                                                  # Number of cylinders
Vsv = np.pi / 4 * bore ** 2 * stroke                                                         # Swept volume
Vts = n_sigma * (np.pi / 4) * (bore ** 2) * stroke                                           # Total swept volume m^3
Vcv = Vsv / (CR -1)                                                                          # Clearance volume m^3
CRt = (Vts + Vcv) / Vcv                                                                      # Trapped compression ratio
Calorific_fuel = 43500000                                                                    # J/kg
Air_fuel_ratio = 14.95                                                                       # Stoichiometric air-fuel ratio for octane
a = 5                                                                                        # Viva Function
m = 2                                                                                        # Viva Function

# INITIAL VALUES FOR TOTAL MASS, AIR MASS, DENSITY, AND DELIVERY RATIO
atmospheric_pressure = 1.12                                                                  # Bar
T1 = atmospheric_temp = 300                                                                  # Kelvin
R = 287                                                                                      # Universal real gas constant J/kgK
max_cylinder_vol = max(df_1500_RPM["VOL(CC)"])
mass_gas1 = atmospheric_pressure * max_cylinder_vol / (R * atmospheric_temp)                 # Mass of gas Kg          
mta = (atmospheric_pressure *100000) * Vsv / (R * atmospheric_temp)                          # Mass of air trapped kg
rho_air_atmos = (atmospheric_pressure*100000) /(R * atmospheric_temp)                        # Air density Kg / m^3
DR = mta / (rho_air_atmos * Vsv)                                                             # Delivery ratio

# HEAT AVAILABLE IN FUEL TO BE RELEASED AT TDC
AFR = 14.7                                                                                   # Trapped air-fuel ratio
mtf = mta / (AFR * 0.9)                                                                      # Mass of fuel trapped Kg
nc = 0.9                                                                                     # Combustion efficiency
# HEAT TRANSFER @ TDC IS EQUIVALENT TO HEAT ENERGY IN FUEL 
Qr = nc * mtf * Calorific_fuel                                                               # Heat released J

# ADIABATIC AND ISENTROPIC COMPRESSION (PROCESS 1-2)
#Pressure @ end of compression
Cv = 718                                                                                     # Specific heat @ constant volume for air J / kgK
Cp = 1007                                                                                    # Specific heat @ constant pressure for air J / kgK
gamma = 1.16                                                                                 # Ratio of specific heat
gamma_1500 = 1.18                                                                            # Ratio of specific heat 
gamma_1750 = 1.19                                                                            # Ratio of specific heat
gamma_2000 = 1.21                                                                            # Ratio of specific heat  
gamma_2500 = 1.20                                                                            # Ratio of specific heat 
gamma_3000 = 1.20                                                                            # Ratio of specific heat
gamma_3500 = 1.21                                                                            # Ratio of specific heat 
gamma_4000 = 1.22                                                                            # Ratio of specific heat 
gamma_4250 = 1.23                                                                            # Ratio of specific heat
gamma_4500 = 1.24                                                                            # Ratio of specific heat
gamma_5000 = 1.23                                                                            # Ratio of specific heat 
gamma_5500 = 1.23                                                                            # Ratio of specific heat
gamma_6000 = 1.23                                                                            # Ratio of specific heat
gamma_6500 = 1.22                                                                            # Ratio of specific heat
gamma_avg = (gamma_1500 + gamma_1750 + gamma_2000 + gamma_2500 + gamma_3000 + gamma_3500 + 
             gamma_4000 + gamma_4250 + gamma_4500 + gamma_5000 + gamma_5500 + gamma_5500 +
             gamma_6000 + gamma_6500 + gamma) / 14
p2 = atmospheric_pressure * (CR) ** gamma                                                    # Pa                                                              
#Temperature @ end of compression
T2 = atmospheric_temp * CR ** (gamma - 1)                                                    # Kelvin 
work_done1_2 = - mass_gas1 * Cv * (T2 - atmospheric_temp)                                    # Joule, Work done during the compression is negative
DeltaU2_1 = mass_gas1 * Cv * (T2 - atmospheric_temp)                                         # Joule, Work done during the compression is positive

# CONSTANT VOLUME COMBUSTION (PROCESS 2-3)
#First law of thermodynamics is applied..
#.. to the combustion process to find the temperature:
T3 = T2 + Qr / (mass_gas1 * Cv)                                                              # Kelvin
p3 = p2 * (T3 / T2)                                                                          # Pascal
DeltaU3_2 = Qr                                                                               # Joule

# CONSTANT VOLUME COMBUSTION (PROCESS 3-4)
T4 = T3 / (CR ** (gamma - 1))                                                                # Kelvin at the end of expasion
p4 = p3 * CR ** (- gamma)                                                                    # pressure at the end of expasion
work_done3_4 = - mass_gas1 * Cv * (T4 - T3)                                                  # Joule, the work done during expansion is positive
DeltaU3_4 = mass_gas1 * Cv * (T4 - T3)                                                       # Joule, the work done during expansion is negative                                                       

# COMPLETE THE CYCLE BY CONSTANT VOLUME HEAT REJECTION (PROCESS 4-1)
DeltaU4_1 =  mass_gas1 * Cv * (atmospheric_temp - T4)                                        # Joule, change of internal energy during heat rejection from 4-1 
#Obtaining Net Values for the Cycle
net_work = work_done1_2 + work_done3_4                                                       # Joule, Net work output from the cycle
Eta_thermal = net_work / Qr                                                                  # Otto cycle thermal efficiency
# Or by:
Eta_thermal_1 = 1 - (1 / CR ** (gamma - 1))                                                  # Otto cycle thermal efficiency
# Must be also difference between heat added and reject
Eta_thermal_2 = (Qr - abs(DeltaU4_1)) / Qr                                                   # Otto cycle thermal efficiency
imep = net_work / Vsv                                                                        # Pascal, indicated mean effective pressure 
Wdot = net_work * (RPM / 120)                                                                # Watts, power output of the engine
mfdot = mtf * (RPM / 120)                                                                    # Kg/s, The fuel consumption rate
isfc_Ws =  mfdot / Wdot                                                                      # Kg/Ws, Specific fuel consumption    
isfc_kWh =  isfc_Ws * 1000 * 3600                                                            # Kg/kWh, Specific fuel consumption    

```

```python
# VALIDATION WITH 3 DIFFERENT ENGINES
#
#                                                                    Honda CBR 1000RR
####################################################################################################################################################
#                                                                    ENGINE PARAMETERS                                                             #
RPM_RANGE_cbr = [*range(1000, 12000, 1000)]                                                        # RPM Range 1000 to 12K                         #                                                                                                                                                                                                 #         
bore_cbr = 0.075                                                                                   # m                                             #
stroke_cbr = 0.0565                                                                                # m                                             #
conn_rod_cbr = 0.10375                                                                             # m                                             #
crank_radius_cbr = stroke_cbr / 2                                                                  # m                                             #
CR_cbr = 12.2                                                                                      # Compression ratio                             #            
n_cbr = 4                                                                                          # Number of cylinders                           #               
densityAir_cbr = 1.201                                                                             # Air density Kg / m^3                          #               
atmospheric_pressure_cbr = 1.12                                                                    # Bar                                           #
InitialTemp_cbr  = 300                                                                             # Kelvin                                        #
Engine_Capacity_cbr = 998.4                                                                        # cc                                            #
Air_fuel_ratio_cbr = 14.95                                                                         # Stoichiometric air-fuel ratio for octane      #
####################################################################################################################################################

################################################## Advancing and retarding ignition timing  ########################################################

###################################
start_combus_cbr = -10 # degree   #
end_combus_cbr  = 25   # degree   #
###################################


################################################################################################################ SIMULATION ##########################################################################################################
Vts_cbr = n_cbr * (np.pi / 4) * (bore_cbr ** 2) * stroke_cbr                                                 # Total swept volume m^3
Vsv_cbr = Vts_cbr / n_cbr                                                                                    # Swept volume
Vcv_cbr = Vsv_cbr / (CR_cbr -1)                                                                              # Clearance volume m^3
CRt_cbr = (Vts_cbr + Vcv_cbr) / Vcv_cbr                                                                      # Trapped compression ratio 
NVolume = 0.90                                                                                               # Vol. Efficiency
R = 287                                                                                                      # Universal real gas constant J/kgK                                                 
mta_cbr = ((atmospheric_pressure_cbr *100000) * Vsv_cbr / (R * InitialTemp_cbr)) * NVolume                   # Mass of gas Kg 
mtf_cbr = mta_cbr / (Air_fuel_ratio_cbr * 0.9)                                                               # Mass of fuel trapped Kg
gamma_cbr = 1.25                                                                                             # Ratio of specific heat

# Creating data for Pressure & Volume
dcbr = {"Crank Angle": [*range(-180, 180 + 1, 1)], "Crank Rad": 0, "Volume cc": 0, "Piston Travel": 0,"Pressure": 0, "dP/Phi": 0}
CBR = pd.DataFrame(data=dcbr)

dres = {"Crank Angle": range(-180,180, 1), "Wiebe": 0, "MBF": 0,"Dqr": 0, "Combustion": 0, "dP/Phi Combustion": 0, "Temperature": 0, "Work Done": 0, "Torque": 0} 
CBR_res = pd.DataFrame(data=dres)

# Converting Angle to Rad. Obtaining piston volume and position.
for k in range(len(CBR["Crank Angle"])):
  CBR["Crank Rad"][k : ] = CBR["Crank Angle"][k] * (np.pi / 180)  # Degrees
  CBR["Piston Travel"][k : ] = (crank_radius_cbr + conn_rod_cbr) - (crank_radius_cbr* np.cos(CBR["Crank Rad"][k])) - (conn_rod_cbr * np.cos(np.arcsin((crank_radius_cbr/conn_rod_cbr) * np.sin(CBR["Crank Rad"][k]) ) ) )  # mm
  CBR["Volume cc"][k : ] = (Vcv_cbr + CBR["Piston Travel"][k] * (np.pi * (bore_cbr / 2) **2)) *1000000  # cc

# Pressure, dP/Phi, with no Combustion.
for i in range(1 , len(CBR["Crank Angle"])):
  CBR["Pressure"][0] = atmospheric_pressure_cbr # Bar
  CBR["Pressure"][i : ] = CBR["Pressure"][i - 1]  * ((CBR["Volume cc"][i - 1] / CBR["Volume cc"][i]) **avg_n_comp) # Bar
  CBR["dP/Phi"][i -1 : ] = CBR["Pressure"][i] - CBR["Pressure"][i -1]

# Viva Function  
for pJ in range(len(CBR_res["Crank Angle"])):
  x = 1 -(np.exp(-5 * ((CBR_res["Crank Angle"][pJ] - (start_combus_cbr )) / (end_combus_cbr - start_combus_cbr))** (2 + 1)))
  # 1000 RPM
  if x < 0:
    x = 0 
  elif x > 0:
    x = x
  CBR_res["Wiebe"][pJ : ] = x

# MFB
for jJ in range(1 , len(CBR_res["Crank Angle"])):
  CBR_res["MBF"][jJ - 1 :] = CBR_res["Wiebe"][jJ] - CBR_res["Wiebe"][jJ - 1]

# Heat Joules
for hT in range(len(CBR_res["Crank Angle"])):
  CBR_res["Dqr"][hT : ] = CBR_res["MBF"][hT] * Calorific_fuel * mtf_cbr * nc

# Combustion Pressure Bar
for cB in range(1 , len(CBR_res["Crank Angle"])):
  CBR_res["Combustion"][0] = CBR["Pressure"][cB]
  CBR_res["Combustion"][cB : ] = (CBR_res["Combustion"][cB - 1] * (CBR["Volume cc"][cB -1] /CBR["Volume cc"][cB] ) ** avg_n_comp)+(1/100000) * (CBR_res["Dqr"][cB]*(gamma_cbr -1)/(CBR["Volume cc"][cB]/1000000))

for dP in range(1 , len(CBR_res["Crank Angle"])):
  CBR_res["dP/Phi Combustion"][0] = 0
  CBR_res["dP/Phi Combustion"][dP - 1 : ] = CBR_res["Combustion"][dP] - CBR_res["Combustion"][dP - 1]

# Temperature K
for tP in range(len(CBR_res["Crank Angle"])):
  CBR_res["Temperature"][tP : ] =((CBR_res["Combustion"][tP] * 100000) * (CBR["Volume cc"][tP] / 1000000))/ ((mta_cbr + mtf_cbr) * R)

# Work Done Joules
for wD in range(1, len(CBR_res["Crank Angle"])):
  CBR_res["Work Done"][wD -1 : ] = (CBR_res["Combustion"][wD - 1] *100000)*((CBR["Volume cc"][wD] - CBR["Volume cc"][wD - 1]) / 1000000)

for Tq in range(1, len(CBR_res["Crank Angle"])):
  CBR_res["Torque"][Tq - 1] = CBR_res["Combustion"][Tq - 1] * (CBR["Volume cc"][Tq] - CBR["Volume cc"][Tq - 1])

# Torque Nm
for Tq in range(1, len(CBR_res["Crank Angle"])):
  CBR_res["Torque"][Tq - 1] = CBR_res["Combustion"][Tq - 1] * (CBR["Volume cc"][Tq] - CBR["Volume cc"][Tq - 1])


```

```python
#                                                                    Honda NC30 
####################################################################################################################################################
#                                                                    ENGINE PARAMETERS                                                             #
RPM_RANGE_vfr = [*range(1000, 12000, 1000)]                                                        # RPM Range 1000 to 12K                         #                                                                                                                                                                                                       
bore_vfr = 0.055                                                                                   # m                                             #
stroke_vfr = 0.0320                                                                                # m                                             #
conn_rod_vfr = 0.11010                                                                             # m                                             #
crank_radius_vfr = stroke_vfr / 2                                                                  # m                                             #
CR_vfr = 11.3                                                                                      # Compression ratio                             #            
n_vfr = 4                                                                                          # Number of cylinders                           #               
densityAir_vfr = 1.201                                                                             # Air density Kg / m^3                          #               
atmospheric_pressure_vfr = 1.12                                                                    # Bar                                           #
InitialTemp_vfr  = 300                                                                             # Kelvin                                        #
Engine_Capacity_vfr = 0.399                                                                        # cc                                            #
Air_fuel_ratio_vfr = 14.95                                                                         # Stoichiometric air-fuel ratio for octane      #
####################################################################################################################################################

################################################## Advancing and retarding ignition timing  ########################################################

###################################
start_combus_vfr = -10 # degree   #
end_combus_vfr  = 25   # degree   #
###################################


################################################################################################################ SIMULATION ##########################################################################################################
Vts_vfr = n_vfr * (np.pi / 4) * (bore_vfr ** 2) * stroke_vfr                                                 # Total swept volume m^3
Vsv_vfr = Vts_vfr / n_vfr                                                                                    # Swept volume
Vcv_vfr = Vsv_vfr / (CR_vfr -1)                                                                              # Clearance volume m^3
CRt_vfr = (Vts_vfr + Vcv_vfr) / Vcv_vfr                                                                      # Trapped compression ratio 
NVolume = 0.90                                                                                               # Vol. Efficiency
R = 287                                                                                                      # Universal real gas constant J/kgK                                                 
mta_vfr = ((atmospheric_pressure_vfr *100000) * Vsv_vfr / (R * InitialTemp_vfr)) * NVolume                   # Mass of gas Kg 
mtf_vfr = mta_vfr / (Air_fuel_ratio_vfr * 0.9)                                                               # Mass of fuel trapped Kg
gamma_vfr = 1.25                                                                                             # Ratio of specific heat

# Creating data for Pressure & Volume
dVF = {"Crank Angle": [*range(-180, 180 + 1, 1)], "Crank Rad": 0, "Volume cc": 0, "Piston Travel": 0,"Pressure": 0, "dP/Phi": 0}
VFR = pd.DataFrame(data=dVF)

dres2 = {"Crank Angle": range(-180,180, 1), "Wiebe": 0, "MBF": 0,"Dqr": 0, "Combustion": 0, "dP/Phi Combustion": 0, "Temperature": 0, "Work Done": 0, "Torque": 0} 
VFR_RES = pd.DataFrame(data=dres2)

# Converting Angle to Rad. Obtaining piston volume and position.
for k in range(len(VFR["Crank Angle"])):
  VFR["Crank Rad"][k : ] = VFR["Crank Angle"][k] * (np.pi / 180)  # Degrees
  VFR["Piston Travel"][k : ] = (crank_radius_vfr + conn_rod_vfr) - (crank_radius_vfr* np.cos(VFR["Crank Rad"][k])) - (conn_rod_vfr * np.cos(np.arcsin((crank_radius_vfr/conn_rod_vfr) * np.sin(VFR["Crank Rad"][k]) ) ) )  # mm
  VFR["Volume cc"][k : ] = (Vcv_vfr + VFR["Piston Travel"][k] * (np.pi * (bore_vfr / 2) **2)) *1000000  # cc

# Pressure, dP/Phi, with no Combustion.
for i in range(1 , len(VFR["Crank Angle"])):
  VFR["Pressure"][0] = atmospheric_pressure_vfr # Bar
  VFR["Pressure"][i : ] = VFR["Pressure"][i - 1]  * ((VFR["Volume cc"][i - 1] / VFR["Volume cc"][i]) **avg_n_comp) # Bar
  VFR["dP/Phi"][i -1 : ] = VFR["Pressure"][i] - VFR["Pressure"][i -1]

# Viva Function  
for pJ in range(len(VFR_RES["Crank Angle"])):
  x = 1 -(np.exp(-5 * ((VFR_RES["Crank Angle"][pJ] - (start_combus_vfr )) / (end_combus_vfr - start_combus_vfr))** (2 + 1)))
  # 1000 RPM
  if x < 0:
    x = 0 
  elif x > 0:
    x = x
  VFR_RES["Wiebe"][pJ : ] = x

# MFB
for jJ in range(1 , len(VFR_RES["Crank Angle"])):
  VFR_RES["MBF"][jJ - 1 :] = VFR_RES["Wiebe"][jJ] - VFR_RES["Wiebe"][jJ - 1]

# Heat Joules
for hT in range(len(VFR_RES["Crank Angle"])):
  VFR_RES["Dqr"][hT : ] = VFR_RES["MBF"][hT] * Calorific_fuel * mtf_vfr * nc

# Combustion Pressure Bar
for cB in range(1 , len(VFR_RES["Crank Angle"])):
  VFR_RES["Combustion"][0] = VFR["Pressure"][cB]
  VFR_RES["Combustion"][cB : ] = (VFR_RES["Combustion"][cB - 1] * (VFR["Volume cc"][cB -1] /VFR["Volume cc"][cB] ) ** avg_n_comp)+(1/100000) * (VFR_RES["Dqr"][cB]*(gamma_vfr -1)/(VFR["Volume cc"][cB]/1000000))

for dP in range(1 , len(VFR_RES["Crank Angle"])):
  VFR_RES["dP/Phi Combustion"][0] = 0
  VFR_RES["dP/Phi Combustion"][dP - 1 : ] = VFR_RES["Combustion"][dP] - VFR_RES["Combustion"][dP - 1]

# Temperature K
for tP in range(len(VFR_RES["Crank Angle"])):
  VFR_RES["Temperature"][tP : ] =((VFR_RES["Combustion"][tP] * 100000) * (VFR["Volume cc"][tP] / 1000000))/ ((mta_vfr + mtf_vfr) * R)

# Work Done Joules
for wD in range(1, len(VFR_RES["Crank Angle"])):
  VFR_RES["Work Done"][wD -1 : ] = (VFR_RES["Combustion"][wD - 1] *100000)*((VFR["Volume cc"][wD] - VFR["Volume cc"][wD - 1]) / 1000000)

for Tq in range(1, len(VFR_RES["Crank Angle"])):
  VFR_RES["Torque"][Tq - 1] = VFR_RES["Combustion"][Tq - 1] * (VFR["Volume cc"][Tq] - VFR["Volume cc"][Tq - 1])

# Torque Nm
for Tq in range(1, len(VFR_RES["Crank Angle"])):
  VFR_RES["Torque"][Tq - 1] = VFR_RES["Combustion"][Tq - 1] * (VFR["Volume cc"][Tq] - VFR["Volume cc"][Tq - 1])
```
